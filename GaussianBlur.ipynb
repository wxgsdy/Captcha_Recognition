```python
import cv2
import os

for filename in os.listdir(r"./captcha/number_low"): #
    if filename.startswith('.'):
        continue
    print (filename)
    imgName =  'captcha/number_low/' + filename
    kernel_size = (5, 5)
    sigma = 1.5
    img = cv2.imread(imgName)
    imgBlur = cv2.GaussianBlur(img, kernel_size, sigma)
    new_imgName = "New_" + str(kernel_size[0]) + "_" + str(sigma) + "_" + filename
    IMG_PATH = os.path.join('./captcha/gausian/', new_imgName)
    print (IMG_PATH)
    cv2.imwrite(IMG_PATH, imgBlur)
```
